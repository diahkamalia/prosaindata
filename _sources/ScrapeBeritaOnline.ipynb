## **Web Scraping**
> **Web Scraping** adalah metode yang digunakan untuk mengumpulkan informasi, baik angka, teks, tabel, maupun media dari berbagai website menggunakan sebuah software yang dapat menyimpan dan mengatur informasi yang telah diunduh.

```python
import requests as req
from bs4 import BeautifulSoup as bs
from datetime import datetime
import csv
hades = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
```

## Keterangan Script
- **import requests as req**, baris kode ini mengimport library requests dengan alias req. Library requests merupakan library yang digunakan untuk mengirim HTTP request ke website yang diinginkan.
- **from bs4 import BeautifulSoup as bs**, Baris kode ini mengimport library BeautifulSoup dengan alias bs dari library bs4. Library BeautifulSoup merupakan library yang digunakan untuk memudahkan proses parsing HTML atau XML.
- **from datetime import datetime**, Baris kode ini mengimport module datetime dari library datetime. Module datetime digunakan untuk menangani tanggal dan waktu.
- **import csv**, baris kode ini mengimport library csv. Library csv merupakan library yang digunakan untuk membaca dan menulis file CSV (Comma Separated Values).
- **hades = {‘user-agent’: ‘Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36’}**, baris kode ini mengisi sebuah dictionary dengan nama hades. Dictionary tersebut berisi informasi user-agent yang akan digunakan saat mengirim HTTP request. Informasi user-agent ini bisa memberitahu website tentang perangkat yang digunakan saat mengirim request[2].

```python
def scrape_CNN(hal):
  global hades
  a = 1
  for page in range(1,hal):
        url = f'https://www.detik.com/search/searchnews?query=world+cup&sortby=time&page={page}'
        ge = req.get(url,hades).text
        sop = bs(ge,'lxml')
        li = sop.find('div',class_='list media_rows list-berita')
        lin = li.find_all('article')
        for x in lin:
            link = x.find('a')['href']
            date = x.find('a').find('span',class_='date').text.replace('WIB','').replace('CNN_News','').split(',')[1]
            headline = x.find('a').find('h2').text
            ge_ = req.get(link,hades).text
            sop_ = bs(ge_,'lxml')
            content = sop_.find_all('div',class_='detail__body-text itp_bodycontent')
            for x in content:
                x = x.find_all('p')
                y  = [y.text for y in x ]
                content_ = ''.join(y).replace('\n', '').replace('ADVERTISEMENT','').replace('SCROLL TO RESUME CONTENT','')
                print(f'done[{a}] > {headline[0:10]}')
                a += 1
                with open('sports.csv','a')as file:
                    wr = csv.writer(file, delimiter=',')
                    wr.writerow([headline,date,link,content_])
```

```python
scrape_CNN(3)
```

## Keterangan Script
- scrape_CNN(3), pemanggilan fungsi scrape_CNN() dengan parameter 3. Fungsi scrape_CNN() merupakan fungsi yang telah dibuat sebelumnya yang berfungsi untuk melakukan scraping data dari website Detik.com dengan kata kunci “world cup”. Parameter 3 yang diberikan merupakan jumlah halaman yang akan di-scrape.
Jadi, jika dijalankan, fungsi scrape_CNN() akan melakukan scraping data dari 3 halaman website Detik.com dengan kata kunci “world cup”. Hasil scraping tersebut kemudian akan disimpan dalam file CSV bernama sports.csv.
